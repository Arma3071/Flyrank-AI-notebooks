## 1. My Lane as an ML Task

My lane is **Machine Learning**.

The task is a **supervised classification** problem. The objective is to predict whether a webpage is likely to become a high-performing page based on historical performance metrics. The model learns patterns from previously observed data and uses them to make predictions for new pages.

This is a supervised learning problem because historical examples are available to train the model.

## 2. Target or Proxy

The target variable is **high_performing_page**, which indicates whether a webpage performs well.

Because there is no direct label for "quality," we use historical performance metrics as a proxy. One possible approach is to label pages with above-median clicks or engagement as high-performing.

Example:

- 1 = High-performing page
- 0 = Not high-performing

```python
import pandas as pd

df = pd.read_csv("/content/flyrank-ml-internship-starter/data/raw/content_refresh_anonymized.csv")

# Example proxy target
df["high_performing_page"] = (
    df["clicks_90d"] > df["clicks_90d"].median()
).astype(int)

df[["clicks_90d", "high_performing_page"]].head()
```

## 3. Success Metric

The primary evaluation metric is **F1 Score**.

F1 Score balances precision and recall, making it appropriate when both correctly identifying high-performing pages and avoiding false predictions are important.

Additional evaluation metrics may include:

- Accuracy
- Precision
- Recall
- ROC-AUC

## 4. Unit of Analysis

Each row in the dataset represents **one webpage**.

The features describe the historical performance of that webpage, such as clicks, impressions, engagement, and other metrics that may help predict future success.

## 5. Why ML Beats a Fixed Rule

A simple rule such as:

"If clicks > 100, then the page is successful"

is too simplistic because webpage performance depends on multiple interacting factors.

Machine learning can simultaneously learn from numerous features such as:

- clicks
- impressions
- engagement rate
- keyword rankings
- historical traffic
- other behavioral signals

Instead of relying on manually selected thresholds, the model automatically discovers relationships within the data and can adapt as new data becomes available.

## 6. Action Supported by the Model

The model's predictions help prioritize webpages for optimization.

For example, if the model predicts that a page has high potential to perform well, the content team can:

- Update the content.
- Improve internal linking.
- Optimize metadata.
- Monitor ranking changes.
- Allocate resources to pages with the greatest expected impact.

This enables data-driven decision-making rather than relying solely on intuition.
